4 结合大模型的使用


```python
from langchain_core.prompts import PromptTemplate

# 创建 PromptTemplate 实例
# input_variables 、 template 是必须要指明的
prompt_template = PromptTemplate(
    input_variables=["role","name"],
    template="你是一个{role}，你的名字是{name}。",
)

print(prompt_template)

# 填充实例中的变量 暂且使用format()方法
prompt = prompt_template.format(role="人工智能专家",name="周夏影")
print(prompt)
```

### 使用from_template()方法创建 PromptTemplate 实例


```python
PromptTemplate.from_template(template="你是一个{role}，你的名字是{name}。")
prompt = prompt_template.format(role="人工智能专家",name="周夏影")
print(prompt)
```

### 两种特殊结构的使用，部分提示词模板的使用、组合提示词模板的使用

```python
## 部分提示词模板的使用

prompt_template = PromptTemplate.from_template(
    template="你是一个{role}，你的名字是{name}。",
    partial_variables={"role":"人工智能专家"}
)
prompt = prompt_template.format(name="周夏影")
print(prompt)
```

```python
## 调用方法partial
prompt_template = PromptTemplate.from_template(
    template="你是一个{role}，你的名字是{name}。")

# partial 调用以后不会对已经存在的变量进行覆盖，需要重新赋值
prompt_template = prompt_template.partial(role="人工智能专家")

prompt = prompt_template.format(name="周夏影")
print(prompt)
```

```python
## 组合提示词的使用
template = PromptTemplate.from_template("给我推荐一些{topic}的电影") + "，时间要求在{time}内"
prompt1 = template.format(topic="动作片",time="1小时")
print(prompt1)
```

### 给变量赋值的两种方式 format()/invoke()

```python
# format(): 直接在 format() 方法中传入变量值
prompt = prompt_template.format(role="人工智能专家",name="周夏影")
print(prompt)
```

```python
# invoke(): 传入一个字典，字典的键是变量名，值是变量值
prompt = prompt_template.invoke({"role":"人工智能专家","name":"周夏影"})
print(prompt)
print(type(prompt))
```

### 调用模型


```python
from langchain_openai import ChatOpenAI
import os
import dotenv

dotenv.load_dotenv(override=True)
print(os.getenv("MODELSPACE_MODEL"))

chat_model = ChatOpenAI(
    model=os.getenv("MODELSPACE_MODEL"), # 默认是gpt-3.5-turbo
    api_key=os.getenv("MODELSPACE_API_KEY"),
    base_url=os.getenv("MODELSPACE_BASE_URL"),
    temperature=0.7,
    # streaming=True # 开启流式调用
)

message = chat_model.invoke(prompt1)
print(message.content)
```
